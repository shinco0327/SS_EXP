## 實驗十
### 傅利葉轉換
### 執行以下程式碼並觀察輸出

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*10*t)
plt.figure(figsize=[10,5])
plt.plot(t, x)
plt.title('Signal plot')

xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)

plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(w_hat, abs(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
from IPython.display import display
import numpy as np
fs = 100
f = 10


lims = (10, 5000, 10)
lims2 = (20, 20000, 20)

def change_freq(wave_freq=10, fs_change=100):
    fs = fs_change
    f = wave_freq
    t = np.arange(0, 1, 1/fs)
    x = np.cos(2*np.pi*f*t)


    plt.figure(1,figsize=[10,5])
    plt.plot(t, x)
    plt.title('Signal plot')

    xf = np.fft.fft(x)
    
    w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
    f = (w_hat/(2*np.pi))*fs
    
    plt.figure(2,figsize=[15,5])

    plt.subplot(121)
    
    plt.plot(f[0:int(fs/2)], abs(xf)[0:int(fs/2)])
    plt.xlabel(r'f (Hz)')
    plt.title('The magnitude spectrum')

    plt.subplot(122)
    plt.plot(np.angle(xf))
    plt.xlabel(r'$\hat\omega$  (radian)')
    plt.title('The phase spectrum')
    print("(5)  類比頻率: "+str(np.argmax(abs(xf))))
    plt.show()
    
w = interactive(change_freq, wave_freq=lims, fs_change=lims2)
display(w)



```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x=np.array([0.,0,0,2,4,6,4,2,0,0,0])

y = x*0

y[0] = 1/3*(x[0])
y[1] = 1/3*(x[1] + x[0])
            
for i in range(2, len(x)):
    y[i] = 1/3*(x[i] + x[i-1] + x[i-2])

plt.stem(y, markerfmt='o', use_line_collection=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 180

t = np.arange(0, 2, 1/fs)

x1 = np.cos(2*np.pi*1*t)
x2 = 0.2*np.cos(2*np.pi*60*t)

x3 = x1+x2

plt.figure(1,figsize=[10,5])
plt.plot(t, x1)
plt.plot(t, x3)
plt.title('x3')

y = x3*0

y[0] = 1/3*(x3[0])
y[1] = 1/3*(x3[1] + x3[0])
            
for i in range(2, len(x3)):
    y[i] = 1/3*(x3[i] + x3[i-1] + x3[i-2])

plt.figure(2,figsize=[10,5])
plt.plot(t, y)
plt.title('y3')
```
