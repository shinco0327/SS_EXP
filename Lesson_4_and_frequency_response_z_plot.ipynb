### (2)上網搜尋scipy.signal.freqz，使用此函式同樣繪製bk={1, 2, 1}之FIR filter之頻率響應。這個函式繪製出來的頻率響應一樣嗎？

```python
#15(1) at slide Page83
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fs = 100
w_hat = np.arange(-np.pi, np.pi, np.pi*2/fs)
H = 1 + 2*np.exp(-1j*w_hat) + 1*np.exp(-1j*w_hat*2) 


plt.figure(figsize=[15,5])
plt.plot(w_hat, abs(H))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The magnitude response')
plt.grid()

print(np.angle(H))
plt.figure(figsize=[15,5])
plt.plot(w_hat[1:], np.angle(H)[1:])
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase response')
plt.grid()
```

```python
#15(2)
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

w_hat = np.arange(-np.pi, np.pi, np.pi*2/fs) 
w_hat, H = signal.freqz([1, 2, 1], 1, w_hat)



plt.figure(figsize=[15,5])
plt.plot(w_hat, abs(H))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The magnitude response')
plt.grid()

plt.figure(figsize=[15,5])
plt.plot(w_hat[1:], np.angle(H)[1:])
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase response')
plt.grid()
```
