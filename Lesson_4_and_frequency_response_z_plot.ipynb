### (2)上網搜尋scipy.signal.freqz，使用此函式同樣繪製bk={1, 2, 1}之FIR filter之頻率響應。這個函式繪製出來的頻率響應一樣嗎？

```python
#15(1)
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fs = 100
w_hat = np.arange(-np.pi, np.pi, np.pi*2/fs)
H = 1 + 2*np.exp(-1j*w_hat) + 1*np.exp(-1j*w_hat*2) 


plt.figure(figsize=[15,5])
plt.plot(w_hat, H)
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The frequency response')
```

```python
#15(2)
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

w_hat = np.arange(-np.pi, np.pi, np.pi*2/fs) 
w_hat, H = signal.freqz([1, 2, 1], 1, w_hat)



plt.figure(figsize=[15,5])
plt.plot(w_hat, H)
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The frequency response')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

angle = np.linspace(-np.pi, np.pi, 50)
cirx = np.sin(angle)
ciry = np.cos(angle)
z = 1
plt.figure(figsize=(8,8))
plt.plot(cirx, ciry,'k-')
plt.plot(np.real(z), np.imag(z), 'o', markersize=12)
plt.plot(0, 0, 'x', markersize=12)
plt.grid()

plt.xlim((-2, 2))
plt.xlabel('Real')
plt.ylim((-2, 2))
plt.ylabel('Imag')
```
